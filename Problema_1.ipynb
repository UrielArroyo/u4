```python
from scipy.stats import norm

# Parámetros dados
media = 70  # media
desviacion_estandar = 4  # desviación estándar

# Definir los límites inferior y superior del intervalo
limite_inferior = 68
limite_superior = 72
```

```python
# Calcular la probabilidad de estudiantes que pesan más de 68 y menos de 72 kg
probabilidad = norm.cdf(limite_superior, media, desviacion_estandar) - norm.cdf(limite_inferior, media, desviacion_estandar)

# Imprimir la probabilidad
print("La probabilidad de estudiantes que pesan más de", limite_inferior, "y menos de", limite_superior, "kg es:", probabilidad)

# Calcular cuántos estudiantes pesan más de 68 y menos de 72 kg (asumiendo una población de 500 estudiantes)
total_estudiantes = 500
estudiantes_en_intervalo = probabilidad * total_estudiantes

# Imprimir el número de estudiantes
print("El número de estudiantes que pesan más de", limite_inferior, "y menos de", limite_superior, "kg es aproximadamente:", round(estudiantes_en_intervalo))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios que sigan una distribución normal con los parámetros dados
np.random.seed(0)  # Para reproducibilidad
datos_pesos = np.random.normal(media, desviacion_estandar, total_estudiantes)

# Graficar el histograma
plt.hist(datos_pesos, bins=30, density=True, color='skyblue', edgecolor='black')

# Añadir líneas verticales para mostrar los límites del intervalo
plt.axvline(x=limite_inferior, color='red', linestyle='--', label='Limite inferior')
plt.axvline(x=limite_superior, color='green', linestyle='--', label='Limite superior')

# Etiquetas y título
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de frecuencia')
plt.title('Distribución de pesos de los estudiantes')

# Mostrar leyenda
plt.legend()

# Mostrar el histograma
plt.show()
```
